# Primera manera

```python
q = quejas_2.groupby(['CLUES','Unidad_involucrada','LATITUD_C','LONGITUD_C','Derechohabientes','Fase',
                      'Dano_agraviado','Tipo_usuario','Motivo','Cierre_acuerdo', 'Causa','Especialidad']).count().reset_index()
```

```python
q.to_excel('prueba_3.xlsx')
```

# Segunda manera

```python
m = quejas_2.groupby(['CLUES','Unidad_involucrada','LATITUD_C','LONGITUD_C','Derechohabientes','Tipo_usuario']).count().reset_index()
```

```python
with pd.ExcelWriter('prueba_2.xlsx', mode='a', engine='openpyxl') as writer:
    m.to_excel(writer, sheet_name='Tipo_usuario')
```
